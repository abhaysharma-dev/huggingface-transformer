```python
from transformers import pipeline
```

```python
clf_model = pipeline("text-classification",model="prithivMLmods/Spam-Bert-Uncased")
```

```python
text = ["URGENT: Your account has been suspended. Click [link] to verify your details immediately to avoid deactivation",
       "CONGRATULATIONS! You have not been selected to win a $1000 gift card.Claim your prize here. ",
       "IRS Final Notice: We are contacting you regarding a tax refund. Call [Number] to claim.",
       "Hey, are we still meeting for lunch at 1 PM?",
        "Your package from Amazon is out for delivery today."]
labels = {"LABEL_1" : "Spam","LABEL_0":"Ham"}
result = clf_model(text)

for res in result:
    label = labels[res["label"]]
    print(label, res["score"])
```
